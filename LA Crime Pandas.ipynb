```python
#dropping columns having more null values 
df=crime.drop(labels=['modus_operandi','weapon_code','weapon_description','crime_code_2','crime_code_3',\
                    'crime_code_4','cross_street'],axis=1)
```

```python
#Changing the datatype of the columns

df['date_reported'] = pd.to_datetime(df['date_reported'])
df['date_occurred'] = pd.to_datetime(df['date_occurred'])
```

```python
df.head(5)
```

```python
#Checking for the statistical analysis
df.describe(include = ['float64','int64'])
```

```python
#From the above descriptive statisticts it is evident that there are negative values in the Victims Age column
```

```python
df['victim_age'].value_counts()
```

```python
# there is Negative values in the Victims age column
df_age = df[df['victim_age']>0]
```

```python
#Adding New Month and Year column in the dataFrame 
df['Year'] = df['date_occurred'].dt.year
df['Month'] = df['date_occurred'].dt.month
```

```python
df
```

## Feature Understanding 

```python
#Month wise Crime count

df['Month'].value_counts().sort_index(ascending=True).plot()
```

```python
df['Year'].value_counts().sort_index(ascending=True).plot(kind='bar')
plt.xticks(rotation=45)
plt.show()
```

```python
# Excluding the month =12 & year =2023

df=df[~((df['Month']== 12)&(df['Year']==2023))]
```

```python
df.sort_values(by='date_occurred')
```

```python
df.groupby(['Year','Month']).size().unstack()
```

```python
crime_over_time = df.groupby(['Year','Month']).size().plot()
crime_over_time 
```

```python
plt.hist(df_age['victim_age'],edgecolor='black',bins=15)
```

```python
#Checking the daily crime rate and doing analysis on this 
df['date_of_occured_no_time'] = df['date_occurred'].dt.date   #adding new column
```

```python
#Per day crime count
crime_count_per_day = df.groupby(['date_of_occured_no_time','crime_description']).size()
crime_count_per_day
```

```python
#Top 20 Crimes
top_20_crimes = crime_count_per_day.groupby(['crime_description']).mean()\
                                   .reset_index(name='average_daily_count').\
                    sort_values(by='average_daily_count',ascending =False).head(20)
```

```python
top_20_crimes
```

```python
top_20_crimes_sorted = top_20_crimes.sort_values(by='average_daily_count',ascending=True)
```

```python
plt.barh(top_20_crimes_sorted['crime_description'],top_20_crimes_sorted['average_daily_count'])
```

```python
#From the above the analysis it is clear that most of the crime is happening is of 'VEHICLE-STOLEN'
```

### Plotting the graphs using subplots 

```python
plt.style.available
```

```python
plt.style.use('Solarize_Light2')
```

```python
plt.figure(figsize=(12,8))

#plot 1- Line Chart 
plt.subplot(2,2,1)
crime_plot = df.groupby(['Year','Month']).size().plot(color='green',linewidth=2,marker='*')
crime_plot.set_xlabel('Year - Month', fontsize=10)
crime_plot.set_title('Crime Over Time')

#Plot 2 - Histogram 
plt.subplot(2,2,2)
plt.hist(df_age['victim_age'],edgecolor='black',bins=15)
plt.tight_layout()
plt.xlabel('Age', fontsize=10)
plt.title('Distribution of Victims Age')

#Plot 3 - Horizontal Bar Chart 
plt.figure()
plt.barh(top_20_crimes_sorted['crime_description'],top_20_crimes_sorted['average_daily_count'],color = 'orange')
plt.title('Top 20 Crimes')
plt.show()
```
